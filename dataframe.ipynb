```python
import numpy as np 
import pandas as pd 
```

**Creating a DataFrame**

```python
# create dataframe using dictonary
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df1=pd.DataFrame(data)
df1
```

```python
# create dataframe using list
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2=pd.DataFrame(data_list)
df2
col=["Name","Age","City","Salary"]
df2=pd.DataFrame(data_list,columns=col)
df2
```

**Selection and Indexing of Columns**

```python
#single column without column name
df2['Age']
```

```python
#single column with column name
df2[['Name']]
```

```python
#multiple column 
df2[['Age','Name']]
```

```python
#single row value
df2.iloc[0]
```

```python
#multigle row value
df2.iloc[[0,1]]
```

**Creating a new column**

```python
df2['Occupation']=['Doctor','Eng','Teacher','Farmer']
df2
```

**Drop**

```python
df2.drop('Occupation',axis = 1,inplace=True)
df2
```

```python
#drop multiple column 
df2.drop(['Name','Age'],axis=1)
```

**Conditional Selection**

```python
#i only want to see people whose age is more than 30
df2[df2['Age']>30]
```

```python
#i only want to see people whose age is more than 30 and their city must be paris
df2[(df2['Age']>30)&(df2['City']=='Paris')]
```
