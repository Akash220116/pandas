```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r"anime.csv")
```

```python
def extract_epi(txt):
    check= False
    data = " "
    for i in txt:
        if i ==')':
            break
        if i =='(':
            check = True
            continue
        if check == True:
            data +=i
    return data
    
            
```

```python
ep=df['Title'].apply(extract_epi)
```

```python
df['episode']= ep.str.replace(" eps"," ").astype(int)
df.head()
```

```python
#which animes have the highest episode count
df[df['episode'] == df['episode'].max()]
```

```python
df['total time'] = df['Title'].str.extract(r'([A-Za-z]{3}\s\d{4}\s-\s[A-Za-z]{3}\s\d{4})')
```

```python
# split start and end
df[['start', 'end']] = df['total time'].str.split(' - ', expand=True)

# convert to datetime
df['start'] = pd.to_datetime(df['start'], format='%b %Y')
df['end'] = pd.to_datetime(df['end'], format='%b %Y')

# calculate month duration (inclusive)
df['month_duration'] = (
    (df['end'].dt.year - df['start'].dt.year) * 12
    + (df['end'].dt.month - df['start'].dt.month)
    + 1
)
```
