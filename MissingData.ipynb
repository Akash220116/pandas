```python
import numpy as np 
import pandas as pd 
```

**Finding Missing Data**

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
pd.isna(df)
```

```python
pd.isna(df).sum()
```

**Filling the missing Data**

```python
#fill all with zero
df.fillna(0)
```

```python
df.fillna(df.mean())
```

```python
#put value for every column
val={'A':15,'B':25,'C':35,'D':45}
df.fillna(val)
```

**Removing Missing Data**

```python
df.dropna() # by default which row has no null value it only exist
```

```python
df.dropna(thresh=3) #  which row has at least 3 value it only show
```
